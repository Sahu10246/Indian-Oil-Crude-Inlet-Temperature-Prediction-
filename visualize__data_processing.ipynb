## density
The density of the crude oil at standard conditions, typically in kg/m¬≥. A key property to assess crude type.

| Crude Type            | Density (kg/m¬≥ @ 15¬∞C) | Notes                                                      |
| --------------------- | ---------------------- | ---------------------------------------------------------- |
| **Light Crude**       | **< 850**              | Low density, high API gravity (> 31.1¬∞), more valuable     |
| **Medium Crude**      | **850 ‚Äì 900**          | API gravity between 22.3¬∞ and 31.1¬∞                        |
| **Heavy Crude**       | **> 900**              | High density, API gravity < 22.3¬∞, harder to refine        |
| **Extra Heavy Crude** | **> 1000**             | Very dense, almost tar-like; API < 10¬∞, requires upgrading |


## API gravity
API Gravity ‚Äì a measure of how heavy or light the crude oil is compared to water. Higher API means lighter crude.

| **API Gravity** | **Crude Type** | **Remarks**                                                                 |
| --------------- | -------------- | --------------------------------------------------------------------------- |
| > 45            | Extra light    | May contain more naphtha, less diesel; less dense, sometimes less efficient |
| 35 ‚Äì 45         | Light          | ‚úÖ **Best for most refiners** ‚Äì high value                                   |
| 22.3 ‚Äì 35       | Medium         | Requires more processing, yields more heavy products                        |
| < 22.3          | Heavy          | Thick, more residuals; less desirable                                       |
| < 10            | Extra heavy    | Difficult to refine; may sink in water                                      |

####Most Valuable (Best) API Range
API Gravity: 35‚Äì45
This range is considered ideal because:
1. It represents light crude oil,
2. High yield of valuable products like gasoline, diesel, and jet fuel.
3.  Easier and cheaper to refine.
4.  Often referred to as "sweet light crude" if sulfur content is also low

## SUL. Wt%
Sulfur content in the crude oil, expressed as weight %. Important for classifying "sweet" or "sour" crude.

| **Property**       | **Ideal Range**    |
| ------------------ | ------------------ |
| **API Gravity**    | **35‚Äì45** (light)  |
| **Sulfur Content** | **< 0.5%** (sweet) |


#### range of sulfur
| **SUL. Wt%**    | **Classification** | **Remarks**                                     |
| --------------- | ------------------ | ----------------------------------------------- |
| **< 0.1%**      | Ultra Sweet        | Excellent, very low sulfur ‚Äì rare & premium     |
| **0.1% ‚Äì 0.5%** | Sweet              | ‚úÖ **Best overall ‚Äì preferred by most refiners** |
| **0.5% ‚Äì 2.0%** | Sour               | Higher processing cost, more emissions          |
| **> 2.0%**      | Very Sour          | Difficult and expensive to process              |

## KV, cSt @ 40C
Kinematic viscosity at 40¬∞C, in centistokes (cSt). Indicates how thick the crude is. Important for pumpability and processing.

It tells you how easily crude oil flows:

1. Higher viscosity = thicker, harder to pump

2. Lower viscosity = thinner, flows more easily

| **KV @ 40¬∞C (cSt)** | **Crude Type**    | **Remarks**                              |
| ------------------- | ----------------- | ---------------------------------------- |
| **< 10 cSt**        | Very Light Crude  | Flows easily; easier to refine           |
| **10 ‚Äì 20 cSt**     | Light Crude       | ‚úÖ Ideal for most refineries              |
| **20 ‚Äì 50 cSt**     | Medium Crude      | Requires more heat/pressure to process   |
| **> 50 cSt**        | Heavy/Extra Heavy | Thick, tar-like; hard to pump and refine |

## CCR
CCR measures the amount of carbon residue left after oil is heated in the absence of oxygen.
It is expressed in weight percent (wt%).

####üîç Why CCR Matters:
Indicates coke-forming tendency

High CCR means:
1. More heavy molecules
2. More likely to produce coke during processing (especially in cracking units)
3. Fouling of refinery equipment

Low CCR means:
1. Cleaner burning
2. Easier to refine

## As
Arsenic is a trace contaminant sometimes found in crude oil, usually in very small amounts (ppb or ppm). Even in tiny quantities, arsenic can be highly problematic in refining.

| Crude Type        | Arsenic Content (ppm) |
| ----------------- | --------------------- |
| Light Sweet Crude | **< 0.1 ppm**         |
| Heavy/Sour Crude  | **0.1 ‚Äì 5 ppm**       |
| High Contaminant  | **> 5 ppm**           |


## crude inlet
Crude Inlet refers to the temperature or pressure (most often temperature) of crude oil entering a processing unit ‚Äî like a desalter, heat exchanger, or furnace ‚Äî in a refinery.

| Processing Unit      | Typical Crude Inlet Temp | Purpose                                              |
| -------------------- | ------------------------ | ---------------------------------------------------- |
| **Desalter**         | 100 ‚Äì 150‚ÄØ¬∞C             | Removes salts & water before heating                 |
| **Preheat Train**    | 150 ‚Äì 300‚ÄØ¬∞C             | Recovers heat from products and raises crude temp    |
| **Furnace (Heater)** | 280 ‚Äì 360‚ÄØ¬∞C             | Final heating before atmospheric distillation column |


‚úÖ Example:
If your data has:

Crude Inlet = 120‚ÄØ¬∞C

Crude Outlet = 310‚ÄØ¬∞C

Then:

It likely passed through desalter + preheat train + furnace

ŒîT = 190‚ÄØ¬∞C ‚Üí used to calculate energy input or heat exchanger performance

## crude outlet
Crude Outlet refers to the temperature (or sometimes pressure) of crude oil leaving a processing unit, such as:

Desalter

Heat exchanger train

Furnace/heater

It‚Äôs usually measured right before the crude enters the atmospheric distillation column.

| Unit/Stage                           | Approx. Crude Outlet Temp | Description                                             |
| ------------------------------------ | ------------------------- | ------------------------------------------------------- |
| **Desalter**                         | 120 ‚Äì 150‚ÄØ¬∞C              | Removes salts and water                                 |
| **Preheat Train (after exchangers)** | 250 ‚Äì 330‚ÄØ¬∞C              | Crude is heated by recovering heat from product streams |
| **Furnace Outlet (to column)**       | 320 ‚Äì 370‚ÄØ¬∞C              | Final heating step before fractionation                 |

## crude flow
Crude Flow refers to the rate at which crude oil flows through a processing unit, such as a desalter, heat exchanger, or furnace in a refinery.

| Unit Size            | Flow Range (Barrels Per Day - BPD) | Flow Range (m¬≥/h) | Description                      |
| -------------------- | ---------------------------------- | ----------------- | -------------------------------- |
| **Small plant/unit** | 5,000 ‚Äì 30,000 BPD                 | 30 ‚Äì 200 m¬≥/h     | Pilot units, mini refineries     |
| **Medium refinery**  | 50,000 ‚Äì 150,000 BPD               | 300 ‚Äì 900 m¬≥/h    | Common in regional refineries    |
| **Large refinery**   | 200,000 ‚Äì 400,000+ BPD             | 1200 ‚Äì 2500+ m¬≥/h | Full-scale industrial refineries |
